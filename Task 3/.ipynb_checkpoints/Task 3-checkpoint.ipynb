```python
#### Here, we see
#### - "Profit" and "Sales" are positively correlated
#### - "Profit" and "Discount" are negatively correlated
```

```python
plt.subplots(figsize=(12,8))
plt.xticks(rotation=90)
sns.countplot(x='State',data=sss_data, order=sss_data['State'].value_counts().index)
```

#### Highest number of buyers are from california and new york

```python
sss_data.groupby("State").sum().sort_values("Sales", ascending=False)
```

#### California has the highest number of Sales

####  North Dakota has the least number of Sales among other states`

```python
sss_data.groupby("State").sum().sort_values("Profit", ascending=False)
```

```python
data_state = sss_data.groupby(['State'])['Sales','Discount','Profit'].sum()
data_state.head()
```

```python
data_state.sort_values('Profit', ascending=False, inplace=True)
data_state[['Profit']].plot(kind = 'bar', figsize=(25,10))
plt.xlabel("Name of States", size = '25')
plt.ylabel("Profit", size = '25')
plt.title("Profit Vs States", size = '25')
plt.legend()
```

####  In terms of Profit we see that Texas is the state in Loss
#### Califronia has the highest number of profits as well

```python
sss_data.groupby("State").sum().sort_values(["Discount"], ascending=False)
```

#### Since Profit and Discount are somewhat correlated , we can see that the state which has the maximum discount is in maximum loss.

```python
sss_data.hist(figsize=(10,10),bins=50)
plt.show()
```

#### Segment wise sales are same in almost all the region

```python
sss_data['Region'].value_counts()
```

```python
data_region = sss_data.groupby(['Region'])['Sales','Discount','Profit'].mean()
data_region.head()
```

#### Discount is given more in the Central region but Profit is more in the West Region

```python
sns.barplot(x='Segment', y='Sales', data = sss_data,hue='Category', order=sss_data['Segment'].value_counts().index, palette='winter_r')
plt.show()
```

#### Sales is more in the Technology Category of the different segments

```python
data_segment = sss_data.groupby(['Segment'])['Sales','Discount','Profit'].mean()
data_segment
```

```python
data_segment.plot.pie(subplots=True, label=data_segment.index, autopct='%1.2f%%', figsize=(20,20))
```

#### '''Sales are more in "Home Office"
#### Discount are more in "Corporate"
#### Profit are more in "Home Office"'''

```python
data_cat = sss_data.groupby(['Category'])['Sales','Discount','Profit'].mean()
data_cat.head()
```

```python
data_cat.plot.pie(subplots=True, label=data_cat.index, autopct='%1.2f%%', figsize=(20,20))
```

#### Highest Sales and Profit is in "Technology"
